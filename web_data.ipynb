```python
import pandas as pd
from urllib.request import urlopen
import json
```

```python
# Settings
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# store the URL in url as parameter for urlopen
url = "https://salty-boy.com/current-match"
  
# store the response of URL
response = urlopen(url)
  
# storing the JSON response from url in data
data_json = json.loads(response.read())
  
# print the json response
print(data_json)
```

```python
# Define columns
columns_match_data = {
    'name': 'name', # 0
    'id': 'id', # 1
    'best_streak': 'best_streak', # 2
    'prev_tier': 'prev_tier', # 3
    'elo': 'elo', # 4
    'tier': 'tier', # 5
    'tier_elo': 'tier_elo', # 6
    'average_bet': 'average_bet', # 7
    'total_matches': 'total_matches', # 8
    'win_rate': 'win_rate', # 9
    'fighter_blue': 'fighter_blue', # 10
    'fighter_red': 'fighter_red', # 11
    'bet_blue': 'bet_blue', # 12
    'bet_red': 'bet_red', # 13
    'match_id': 'id', # 14
    'match_format': 'match_format', # 15
    'streak_blue': 'streak_blue', # 16
    'streak_red': 'streak_red', # 17
    'match_tier': 'tier', # 18
    'winner': 'winner' # 19
}

data_extraction = list(columns_match_data.values())

# Define DataFrame
match_data = pd.DataFrame()

# Define lists for data extraction
sides = ['fighter_blue_info', 'fighter_red_info']

for side in sides:
    if data_json[side] == None:
        continue
    for idx, match in enumerate(data_json[side]['matches']):
        temp_data = [
            data_json[side][data_extraction[0]],
            data_json[side][data_extraction[1]],
            data_json[side][data_extraction[2]],
            data_json[side][data_extraction[3]],
            data_json[side][data_extraction[4]],
            data_json[side][data_extraction[5]],
            data_json[side][data_extraction[6]],
            data_json[side]['stats'][data_extraction[7]],
            data_json[side]['stats'][data_extraction[8]],
            data_json[side]['stats'][data_extraction[9]],
            data_json[side]['matches'][idx][data_extraction[10]],
            data_json[side]['matches'][idx][data_extraction[11]],
            data_json[side]['matches'][idx][data_extraction[12]],
            data_json[side]['matches'][idx][data_extraction[13]],
            data_json[side]['matches'][idx][data_extraction[14]],
            data_json[side]['matches'][idx][data_extraction[15]],
            data_json[side]['matches'][idx][data_extraction[16]],
            data_json[side]['matches'][idx][data_extraction[17]],
            data_json[side]['matches'][idx][data_extraction[18]],
            data_json[side]['matches'][idx][data_extraction[19]],
        ]
        match_data = pd.concat([match_data, pd.DataFrame([temp_data], columns = list(columns_match_data.keys()))])
```
