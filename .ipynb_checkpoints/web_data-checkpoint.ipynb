```python
# Libs
import pandas as pd
from urllib.request import urlopen
import json
```

```python
# Functions

def get_fighter_data(fighter_id):
    # returns the data for the specified fighter
    response_fighter = urlopen(fighter_url+str(fighter_id))
    fighter_data_json = json.loads(response_fighter.read())
    
    columns_fighter = {
        'name': 'name', # 0
        'id': 'id', # 1
        'best_streak': 'best_streak', # 2
        'prev_tier': 'prev_tier', # 3
        'elo': 'elo', # 4
        'tier': 'tier', # 5
        'tier_elo': 'tier_elo', # 6
        'average_bet': 'average_bet', # 7
        'total_matches': 'total_matches', # 8
        'win_rate': 'win_rate', # 9
    }

    data_extraction_fighters = list(columns_fighter.keys())

    for key in columns_fighter_data:
        temp = [
            fighter_data_json[data_extraction_fighters[0]],
            fighter_data_json[data_extraction_fighters[1]],
            fighter_data_json[data_extraction_fighters[2]],
            fighter_data_json[data_extraction_fighters[3]],
            fighter_data_json[data_extraction_fighters[4]],
            fighter_data_json[data_extraction_fighters[5]],
            fighter_data_json[data_extraction_fighters[6]],
            fighter_data_json['stats'][data_extraction_fighters[7]],
            fighter_data_json['stats'][data_extraction_fighters[8]],
            fighter_data_json['stats'][data_extraction_fighters[9]]
        ]
    return temp

def get_winner(row):
    # returns 0 if the fighter won the match, and 1 otherwise
    if row.id == row.winner:
        return 0
    else:
        return 1

def get_home_other_fighter(row):
    # returns the name of the other fighter in the match
    if row.id == row.fighter_blue:
        return row.fighter_red
    else:
        return row.fighter_blue

def update_fighter_data(df, name_other, away_fighter):
#     print([name_other, away_fighter])
    # updates the data for the specified fighter
    row_index = df.index[df['name_other'] == name_other].tolist()[0]
    if name_other == -421:
        df.iloc[row_index, 22:] = get_fighter_data(away_fighter)
#         df.loc[row_index, 22:] = get_fighter_data(away_fighter)

def update_all_fighter_data(df):
    # updates the data for all fighters in the dataframe
    for i, row in df.iterrows():
        update_fighter_data(df, row['name_other'], row['away'])

def process_match_data(match_data):
    # set up columns for fighter data
    columns_fighter_data = {
        'name_other': 'name', 
        'id_other': 'id', 
        'best_streak_other': 'best_streak', 
        'prev_tier_other': 'prev_tier', 
        'elo_other': 'elo', 
        'tier_other': 'tier', 
        'tier_elo_other': 'tier_elo', 
        'average_bet_other': 'average_bet', 
        'total_matches_other': 'total_matches', 
        'win_rate_other': 'win_rate',
    }

    fighter_other_names = list(columns_fighter_data.keys())

    for idx in fighter_other_names:
        match_data[idx] = -421
        
    # adds columns to the match data dataframe
    match_data['winner_binary'] = match_data.apply(get_winner, axis=1)
    match_data['away'] = match_data.apply(get_home_other_fighter, axis=1)
    update_all_fighter_data(match_data)
    return match_data
```
